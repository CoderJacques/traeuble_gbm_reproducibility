```python
import pandas as pd
import numpy as np
import scanpy as sc
```

```python
sc_data = sc.read_h5ad("mofa/mofa_workflow/input_data_mosaic/sc_mosaic_rounded_embeds_NOuncert_subclusters_mofa.h5ad")
```

```python
sc_samples = sc_data.obs["sample_id"].values.unique()
```

```python
clinical = pd.read_csv("mofa/mofa_workflow/input_data_mosaic/Prepared_Sample_Meta_Data.csv")
```

```python
remove = ["HKG056a" ,"HKG106a", 'HKG017b', 'HKG029b', 'HKG036b', 'HKG044b', 'HKG059b', 'HKG061b','HKG079b', 'HKG082b', 'HKG084b', 'HKG085a', 'HKG090b', 'HKG091a',
'HKG096b', 'HKG100b', 'HKG105b', 'HKG109b', 'HKG111b', 'HKG113b','HKG115b'] # recurrent samples that are not included in the study except for WES comparisons
```

```python
baseline = clinical[~clinical['sample_id'].isin(remove)]
```

```python
baseline_sc = baseline[baseline['sample_id'].isin(sc_samples)] # keep only samples that are in the sc dataset because the rest was not used as input in the mofa model
```
